```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = plt.imread('images/multipleeyes.jfif')
plt.imshow(img)
# to display the image now detect the eyes in the image
```

```python
model = cv2.CascadeClassifier('haarcascade_eye.xml')
eyes = model.detectMultiScale(img)
```

```python
eyes
```

```python
for eye in eyes:
    print(eye)
    x,y,w,h = eye
    img = cv2.rectangle(img,(x,y),(x+w,y+h), (0,0,0),4)
```

```python
plt.imshow(img)
```
