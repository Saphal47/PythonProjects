```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = plt.imread('images/fullbody.jfif')
plt.imshow(img)
```

```python
model = cv2.CascadeClassifier('haarcascade_fullbody.xml')
body = model.detectMultiScale(img)
```

```python
body
```

```python
for bod in body:
    x,y,w,h = bod
    img = cv2.rectangle(img,(x,y),(x+w,y+h), (255,0,0),4)
    plt.imshow(img)
```
